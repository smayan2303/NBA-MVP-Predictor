```python
import requests
import os
import shutil
import time
```

```python
years = list(range(1991, 2024))
url_base = "https://www.basketball-reference.com/awards/awards_{}.html"
```

```python
for year in years: 
    url = url_base.format(year)
    data = requests.get(url)
    with open("mvps/{}.html".format(year), "w+", encoding = "utf-8") as f:
        f.write(data.text)
```

```python
from bs4 import BeautifulSoup
import pandas as pd
from io import StringIO
```

```python
#Testing Opening Files

with open("mvps/1991.html", encoding = "utf-8") as f:
        page = f.read()
soup = BeautifulSoup(page, 'html.parser')
soup.find('tr', class_="over_header").decompose()
mvp_table = soup.find_all(id="mvp")[0]
tables = pd.read_html(StringIO(str(mvp_table)))
mvp_1991 = tables[0]
mvp_1991
```

```python
#ONLY RUN AFTER WEBSCRAPING IS SUCCESSFUL
dfs = []
for year in years:
    with open("mvps/{}.html".format(year), encoding = "utf-8") as f:
        page = f.read()

    soup = BeautifulSoup(page, 'html.parser')
    soup.find('tr', class_="over_header").decompose()
    mvp_table = soup.find_all(id="mvp")[0]
    mvp_df = pd.read_html(StringIO(str(mvp_table)))[0]
    mvp_df["Year"] = year
    dfs.append(mvp_df)
```

```python
mvps = pd.concat(dfs)
mvps.tail()
```

```python
mvps.to_csv("mvps.csv")
```

```python
from selenium import webdriver
from selenium.webdriver.common.keys import Keys
from selenium.webdriver.chrome.service import Service
```

```python
service = Service(executable_path="./chromedriver.exe")
driver = webdriver.Chrome(service=service)
```

```python
for year in years:
    url = url_players.format(year)

    driver.get(url)
    driver.execute_script("window.scrollTo(1, 5000)")
    time.sleep(2)
    html = driver.page_source
    with open("players/{}.html".format(year), "w+", encoding = "utf-8") as f:
        f.write(html)
```

```python
dfs = []
for year in years:
    with open("players/{}.html".format(year), encoding = "utf-8") as f:
        page = f.read()
    soup = BeautifulSoup(page, 'html.parser')
    soup.find('tr', class_="thead").decompose()
    player_table = soup.find_all(id="per_game_stats")[0]
    player_df = pd.read_html(StringIO(str(player_table)))[0]
    player_df["Year"] = year
    dfs.append(player_df)
```

```python
players = pd.concat(dfs)
players.head()
```

```python
players.to_csv("players.csv")
```

```python
url_teams = "https://www.basketball-reference.com/leagues/NBA_{}_standings.html"
for year in years:
    url = url_teams.format(year)
    data = requests.get(url)
    with open("teams/{}.html".format(year), "w+", encoding = "utf-8") as f:
        f.write(data.text)
```

```python
dfs = []
for year in years:
    with open("teams/{}.html".format(year), encoding = "utf-8") as f:
        page = f.read()
    soup = BeautifulSoup(page, 'html.parser')
    soup.find('tr', class_="thead").decompose()
    e_table = soup.find_all(id="divs_standings_E")[0]
    e_df = pd.read_html(StringIO(str(e_table)))[0]
    e_df["Year"] = year
    e_df["Team"] = e_df["Eastern Conference"]
    del e_df["Eastern Conference"]
    dfs.append(e_df)
    
    w_table = soup.find_all(id="divs_standings_W")[0]
    w_df = pd.read_html(StringIO(str(w_table)))[0]
    w_df["Year"] = year
    w_df["Team"] = w_df["Western Conference"]
    del w_df["Western Conference"]
    dfs.append(w_df)
```

```python
teams = pd.concat(dfs)
teams.head()
```

```python
teams.to_csv("teams.csv")
```
